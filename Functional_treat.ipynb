```
data = []
dataset_summary = { }
```

```
def factorial(value): 
    if value == 0 or value == 1:
        return 1
    return value * factorial(value-1)
```

```
def data_summary():
    """return maximum value, minimum value, total of the list values"""
    print("----------- Data Summary -----------")
    if not data: 
        return None, None, None

    #print max value from the list
    max_val = max(max(r) for r in data)
            
    #print min value from the list
    min_val = min(min(r) for r in data)

    #print sum of all the value from the list
    total = sum(sum(r) for r in data)

    return max_val, min_val, total
```

```
def compute_summary():
    """Compute total elements and mean using global variable."""
    global dataset_summary

    if not data:
        print("No data available!")
        return

    total = sum(sum(r) for r in data)
    count = sum(len(r) for r in data)

    dataset_summary = {
        "total_elements": count,
        "Average": total / count
    }
```

```
def display_args(*args):
    """display multiple values using *args."""
    print("values using *args: ")
    labels = ["Min", "Max", "Total"]
    for label, val in zip(labels, args):
        print(f"{label}: {val}")
```

```
def display_kwargs(**kwargs):
    """display dataset summary using **kwargs."""
    for key, value in kwargs.items():
        print(f"{key} : {value}")
```

```
def filter_data(threshold):
    """Filter 2D array values greater than given threshold value using lambda and filter function"""
    if not data:
        print("No Data Available!!!!")
        return
    
    filtered = []

    filtered = list(filter(lambda x: x> threshold, [val for row in data for val in row])) 
    print("Filtered Data: ", filtered)
```

```
def sort_data():
    """Sort 2D data using sorted()"""
    if not data:
        print("No Data Available!!!")
        return
    
    print("Choose Sorting option:")
    print("1. Ascending")
    print("2. Descending")
    choice = int(input("Enter your choice: "))

    flat = [val for row in data for val in row]

    if choice == 1:
        sorted_data = sorted(flat)
        print("Sorted Data in Ascending order: ", sort_data)
    elif choice == 2:
        sorted_data = sorted(flat, reverse=True)
        print("sorted Data in Decending order: ", sort_data)
    else:
        print("Invalid Choice!!!")
        return    
    print(", ".join(map(str, sorted_data)))
```

```
while True: 
    print("----------- Main Menu -----------\n" \
    "1. Input Data \n" \
    "2. Display Data Summary \n" \
    "3. Calculate Factorial \n" \
    "4. Filter Data by Threshold \n" \
    "5. Sort Data \n" \
    "6. Display Dataset Statistics \n" \
    "7. Exit Program")

    choice = int(input("Please Enter your choice: "))

    if choice == 1: 
        print("----------- Input Data -----------")
        #getting data from user for 2D array
        row = int(input("please enter number of rows for array: "))
        col = int(input("please enter number of column of array: "))

        for r in range(row):
            row_data = []
            for c in range(col):
                col_value = int(input(f"please enter {c+1} column value: "))
                row_data.append(col_value)
            data.append(row_data)
    
    elif choice == 2:
        for r in data:
            print(r)
        print(compute_summary.__doc__)
        compute_summary()

        if dataset_summary:
            display_kwargs(**dataset_summary)


    elif choice == 3: 
        value = int(input("Please enter number to calulate its factorial: "))
        print(f"Factorial of the {value} is: ",factorial(value))

    elif choice == 4:
        print(filter_data.__doc__)   # using __doc__
        threshold = int(input("Enter threshold value: "))
        filter_data(threshold)

    elif choice == 5:
        print(sort_data.__doc__)   # using __doc__
        sort_data() 
    
    elif choice == 6:
        print(data_summary.__doc__)

        max_val, min_val, total = data_summary()

        display_args(min_val,max_val,total)
        
        compute_summary()
        display_kwargs(**dataset_summary)
    
    elif choice == 7: 
        print("Exiting the program.....")
        break

    else:
        print("Invalid Choice!!")
```
